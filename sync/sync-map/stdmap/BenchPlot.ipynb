```python
import pandas as pd

# go test -bench . | grep ^Benchmark | awk '{print $1"\t"$3}'
df = pd.read_csv("data.tsv", sep="\t", header=None)
data = {
    "builtin": df[~df[0].str.contains("Sync")].iloc[:, 1].values,
    "sync": df[df[0].str.contains("Sync")].iloc[:, 1].values,
}
index = ["store", "delete", "load"]
ax = pd.DataFrame(data, index=index).plot(kind='bar',
                                     figsize=(10, 6),
                                     title='Go Benchmark builtin map and sync.Map (04/2020)')
ax.set_ylabel("ns/op")
ax.set_xticklabels(index, rotation=45);
```
